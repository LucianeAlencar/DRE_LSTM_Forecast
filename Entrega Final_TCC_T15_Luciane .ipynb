#Tratamento das variaveis Categoricas. 
 por definição o tratamento das variaveis categoricas, é feito considerando a cardinalidade.

 para cardinalidade > 10 faremos label encoder
 para cardinalidade <=10 faremos get dummies

```python
df_escolhidas_estat.nunique()
```

```python
#tratando as Dumies de acordo com a classificação acima = DENOM_CIA; DT_INI_EXERC; DT_FIM_EXERC
df_tratados = pd.get_dummies(df_escolhidas_estat,
                      columns=['DENOM_CIA',
                               'DT_INI_EXERC',
                               'DT_FIM_EXERC'],
                      drop_first=True,
                      prefix=['DENOM_CIA',
                               'DT_INI_EXERC',
                               'DT_FIM_EXERC'],
                      prefix_sep='*')
df_tratados.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import(model_selection, preprocessing)
```

```python
#separando as variaveis explicativas
explicativas  = df_tratados.drop(columns=['Lucro do Periodo'], axis=1)
target = df_tratados['Lucro do Periodo']
```

```python
explicativas
```

```python
#para separar a base em treino e teste
from sklearn.model_selection import train_test_split
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(explicativas,
                                                       target,
                                                       test_size=0.3,
                                                       random_state=42)
```

```python
x_treino.shape
```

```python
x_teste.shape
```

```python
# import do modulo dentro do sklearn
from sklearn.linear_model import LinearRegression
```

```python
# chamada do objetivo
linear = LinearRegression()
```

```python
# treinamento
linear.fit(x_treino, y_treino)
```

```python
# resultado

from sklearn.metrics import r2_score
```

```python
r2_score(y_treino, linear.predict(x_treino))
```

```python
r2_score(y_teste, linear.predict(x_teste))
```

```python
linear.intercept_
```

```python
# conversão de series para aprendizado supervisionado
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
	n_vars = 1 if type(data) is list else data.shape[1]
	df = pd.DataFrame(data)
	cols, names = list(), list()
	# sequencia (t-n, ... t-1)
	for i in range(n_in, 0, -1):
		cols.append(df.shift(i))
		names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
	# forecast da sequencia (t, t+1, ... t+n)
	for i in range(0, n_out):
		cols.append(df.shift(-i))
		if i == 0:
			names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
		else:
			names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]

	agg = pd.concat(cols, axis=1)
	agg.columns = names
	# drop rows with NaN values
	if dropnan:
		agg.dropna(inplace=True)
	return agg
```
